```python
# Bar graph of diagnosis
sns.countplot(x='diagnosis', data=df)
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()

# Pie chart of diagnosis
labels = df['diagnosis'].unique()
sizes = df['diagnosis'].value_counts()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Diagnosis Distribution')
plt.show()
```

```python
# now encode the diagnosis column values
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
# Correlation
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(), annot=True, fmt='.0%')
```

```python
# Checking the Skewness of the columns

# Check for normality using Shapiro-Wilk test
from scipy.stats import shapiro
for col in df.columns:
  _, p_value = shapiro(df[col])
  if p_value < 0.05:
    print(f"{col} is not normally distributed.")
  else:
    print(f"{col} is normally distributed.")

# Draw a graph individually for each column
for col in df.columns:
  plt.figure()
  sns.histplot(data=df, x=col, kde=True)
  plt.title(f"Distribution of {col}")
  plt.show()
```

```python
# now checking the outliers in each column

# Calculate the interquartile range (IQR) for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers based on the IQR
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Create a boxplot for each column
for col in df.columns:
  plt.figure()
  sns.boxplot(data=df, x=col)
  plt.title(f"Boxplot of {col}")
  plt.show()
```

```python
# Preprocessing like minmaxscaling(for fixed range), Robust scaling (for outlier handling)
#Next step__ Column Transformer, Pipline (to build chain), Data Training

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Define the columns to be normalized and the columns to have outliers removed
numeric_features = df.columns.drop('diagnosis')
categorical_features = []

# Create a ColumnTransformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('minmax', MinMaxScaler()),
            ('robust', RobustScaler())
        ]), numeric_features)
    ])

# Create a pipeline that includes the preprocessor and a Logistic Regression model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop('diagnosis', axis=1), df['diagnosis'], test_size=0.2, random_state=42)

# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)

# Evaluate the model on the testing data
score = pipeline.score(X_test, y_test)
print('Test accuracy:', score*100)
```

```python
# Hyperparamete tunning

# Import GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
params = {
    'classifier__C': [0.1, 0.5, 1.0, 2.0, 5.0],
    'classifier__penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'classifier__solver': ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
    'classifier__max_iter': [100, 200, 300, 400, 500]
}

# Create a GridSearchCV object
grid_search = GridSearchCV(pipeline, params, cv=5, n_jobs=-1, verbose=1)

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and score
print('Best parameters:', grid_search.best_params_)
print('Best score:', grid_search.best_score_)
```

```python
# Now export model using pickel

import pickle

# Save the trained model to a file
with open('model.pkl', 'wb') as f:
  pickle.dump(grid_search.best_estimator_, f)

```
