```python
#import Libaraies
import pandas as pd
import numpy as np
import datetime
```

```python
#import data sets
'''
Remember Orders data set is used for date type data handling   and messages data set is 
used for time handling and performed differnt tasks on these datasets
'''
df_date= pd.read_csv('orders.csv')
df_time= pd.read_csv('messages.csv')
```

```python
df_date.head(5)
```

```python
df_date.info()
```

```python
# Changing the data type of column (date) in dataset
df_date['date'] = pd.to_datetime(df_date['date'])
```

```python
df_date.info()
```

```python
#Extracting day from date column as you can see following  
df_date['day'] = df_date['date'].dt.day
df_date
```

```python
# Extracting month from dataset
df_date['month'] = df_date['date'].dt.month
df_date
```

```python
# Extracting year from dataset
df_date['year'] = df_date['date'].dt.year
df_date
```

```python
# Extracting day of week from dataset

df_date['day_of_week'] = df_date['date'].dt.dayofweek  # Monday is 0 and Sunday is 6
df_date
```

```python
# Extracting quarter from dataset

df_date['quarter'] = df_date['date'].dt.quarter
df_date
```

```python
# Extracting decade from dataset

df_date['decade'] = (df_date['date'].dt.year // 10) * 10
df_date
```

```python
# Extracting semester from dataset

df_date['semester'] = df_date['date'].dt.quarter.map({1: '1st Semester', 2: '2nd Semester', 3: '3rd Semester', 4: '4th Semester'})
df_date
```

```python
# Extracting weekend from dataset

df_date['weekend'] = df_date['date'].dt.weekday.isin([5, 6])  # True if Saturday or Sunday, False otherwise
df_date
```

```python
# Extracting day name from dataset

df_date['day_name'] = df_date['date'].dt.day_name()
df_date
```

```python
# Extracting number of week from dataset

df_date['date_week'] = df_date['date'].dt.week
df_date
```

```python
# Now importing today date and time

today = datetime.datetime.today()

today
```

```python
# Extracting month from dataset  and subtracting from This month mean recent month

np.round((today - df_date['date']) / np.timedelta64(1, 'M'),0)
```

<h1> Time handling </h1>

```python
# Exploring dataset for time example
df_time
```

```python
df_time.info()
```

```python
df_time['date'] = pd.to_datetime(df_time['date'])
```

```python
df_time.info()
```

```python
# Extracting hour, minute, second from dataset  


df_time['hour'] = df_time['date'].dt.hour
df_time['min'] = df_time['date'].dt.minute
df_time['sec'] = df_time['date'].dt.second
df_time
```

```python
# separating the time from date column

df_time['time'] = df_time['date'].dt.time
```

```python
df_time
```

```python
# Extracting days from dataset  and subtracting from current day according the recent time

today - df_time['date']
```

```python
(today - df_time['date']).dt.days
```

```python
# Extracting hours from dataset  and subtracting from current hour

(today - df_time['date'])/np.timedelta64(1,'h')
```
