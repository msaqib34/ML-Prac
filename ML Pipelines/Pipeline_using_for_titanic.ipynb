```python
import pandas as pd
import numpy as np
```

```python
# import imputer , one hot encoding , make pipe line and column transformer ,and decision tree  and import test split train, and import min max scaling, and accuracy test

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv('/content/train.csv')
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'], inplace=True)
```

```python
df
```

```python
data_train = train_test_split
X_train,x_test,y_train,y_test= data_train(df.drop(columns=['Survived']), df['Survived'], test_size=0.2 , random_state=42 )
```

```python
trf1 = ColumnTransformer(
    [
        ('age', SimpleImputer(), [2]),
        ('Embarked', SimpleImputer(strategy='most_frequent'),[6])
    ],remainder='passthrough'
)
```

```python
trf2= ColumnTransformer(
    [
       ('ohe_sex_embarked',OneHotEncoder(sparse=False , handle_unknown='ignore'), [1,6])
    ]
)
```

```python
trf3= ColumnTransformer(
    [
      ( 'scalar', MinMaxScaler(), slice(0,10))
    ]
)
```

```python
trf4= DecisionTreeClassifier()
```

```python
pipe=Pipeline(
    [
        ('trf1',trf1),
        ('trf2',trf2),
        ('trf3', trf3),
        ('trf4', trf4 )
    ]
)
```

```python
pipe.fit(X_train, y_train)
```

```python
testing=pipe.predict(x_test)
```

```python
accuracy_score(y_test,testing)
```

```python
# prompt: do cross validation

from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=10).mean()
```
